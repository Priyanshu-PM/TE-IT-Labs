```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.cluster import AgglomerativeClustering

import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('cities_r2.csv')
data.info()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
xx = data[['effective_literacy_rate_total']]
```

```python
sc = StandardScaler()

x = sc.fit_transform(xx)
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 8))
plt.title("Dendrogram", fontsize=20)
plt.xlabel("Cities")
plt.ylabel("Distance")

dend = shc.dendrogram(shc.linkage(x, method='ward'))
```

```python
# from above figure, we can conclude that the number of clusters is 3
```

```python
cluster = AgglomerativeClustering(n_clusters = 3, affinity= 'euclidean', linkage = 'ward')
```

```python
pred = cluster.fit_predict(x)
```

```python
pred
```

```python
data['cluster'] = pred
```

```python
data.head()
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(xx, xx.index, c=data['cluster'], cmap='rainbow')
plt.title('Aglomerative Clustering', fontsize=20)
plt.xlabel('Cities', fontsize=16)
plt.ylabel('Effective Literacy Rate', fontsize=16)
plt.show()
```
